```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
import pandas as pd
data=pd.read_csv(r"/content/diabetes.csv")
data.head()
```

```python
aml = H2OAutoML(max_models =25,
                balance_classes=True,
                seed = 1)
```

```python
train_h2o=h2o.H2OFrame(data)
```

```python
x=data.drop("Outcome",axis=1).columns
y="Outcome"
```

```python
aml=h2o.automl.H2OAutoML(max_models=25,balance_classes=True,seed=42)
```

```python
aml.train(x=list(x), y=y, training_frame=train_h2o)
```

```python
lb=aml.leaderboard
```

```python
best_model = aml.get_best_model()
print(best_model)
```
